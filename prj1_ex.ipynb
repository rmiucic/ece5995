```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import sys
from gps_pkg import gps as gp
```

```python
df = pd.read_csv('test.csv')
```

```python
df
```

```python
df['distance']=0.0
```

```python
df
```

```python
for ix in range(len(df)):
    if ix > 0:
        #print('longitude:'+str(df.longitude[ix]))
        #print('latitude:'+str(df.latitude[ix]))
        d = gp.GPSDistance(df.latitude[ix],df.longitude[ix],df.latitude[ix-1],df.longitude[ix-1])
        #print(d)
        df['distance'][ix]=d
        
```

```python
df
```

```python
df['speed']=0.0
```

```python
df
```

```python
time_you_figure_out=2
for ix in range(len(df)):
    if ix > 0:
        df['speed'][ix]=df['distance'][ix]/time_you_figure_out
```
